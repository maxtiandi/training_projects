## Предобработка данных

```python
# Напишем функцию для перевода названия столбцов из CamelCase в snake_case для удобства
def to_snake_case(columns):
    new_cols = []
    for name in columns:
        name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
        name = re.sub('__([A-Z])', r'_\1', name)
        name = re.sub('([a-z0-9])([A-Z])', r'\1_\2', name).lower()
        new_cols.append(name)
    return new_cols
```

```python
# Исправим названия столбцов на более удобные
data_autos.columns = to_snake_case(data_autos.columns)
# Проверка 
data_autos.head()
```

```python
# Пайплайн для замены пропущенных значений
imputer_pipe = Pipeline(
    [
        (
            'SimpleImputer',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
    ]
)

# Обработаем пропущенные значения при помощи пайплайна
nan_cols = ['vehicle_type', 'gearbox', 'model', 'fuel_type', 'repaired']
data_autos[nan_cols] = imputer_pipe.fit_transform(data_autos[nan_cols])

# Проверка
print(f'Количество пропущенных значений после замены: \n {data_autos.isna().sum()}')
print('')
print(f'Проверка исправности данных: {data_autos.info()}')
```

- Подобных записей в данных весьма много *(11.32%)*: предлагаю заменить нули на медианное значение количества лошадиных сил среди подобных брендов и моделей автомобиля. 
- В том числе являются странными записи, в которых в автомобилях более 1500 л.с., а тем более 10 000 л.с. Подобные авто либо являются суперболидами (а суперболиды вряд ли будут продавать на платформе по продаже авто с пробегом), либо это техническая ошибка сервиса. Таких записей меньше - *0.056%*, однако предлагаю заменить подобные значения по тому же принципу


---

```python
# Заменим нули и нужные значения в столбце 'power' на NaN, чтобы они не учитывались при вычислении медианы
data_autos['power'].replace(0, pd.NA, inplace=True)
data_autos.loc[(data_autos['power'] > 1500), 'power'] = pd.NA

data_autos['power'] = pd.to_numeric(data_autos['power'], errors='coerce')

# Вычисляем медиану для каждой уникальной пары 'brand' и 'model'
median_power = data_autos.groupby(['brand', 'model'])['power'].median().reset_index()

# Объединяем данные по медианам с основным датасетом по 'brand' и 'model'
data_autos = pd.merge(data_autos, median_power, on=['brand', 'model'], how='left', suffixes=('', '_median'))

# Проверка объединения
print('Датасет ДО изменения, с дополнительным столбцом: \n')
display(data_autos.head())
display(data_autos.shape)
display(data_autos.info())

# Замена значений 'power' на медианные значения, если исходное значение было 0 или больше 1500
data_autos['power'] = data_autos.apply(lambda row: row['power_median'] if pd.isna(row['power']) else row['power'],
                                       axis=1)

# Удаление вспомогательной колонки с медианными значениями
data_autos.drop(columns=['power_median'], inplace=True)

print('----------')
print(' ')
print('Датасет ПОСЛЕ изменения: \n')
display(data_autos.head())
display(data_autos.shape)
display(data_autos.info())
```

## Подготовка выборок для обучения

---
Предлагаю избавиться от следующих столбцов, которые не принесут особой пользы для обучения:
- `number_of_pictures` - в признаке только нули, это скорее большая техническая ошибка. Фича бесполезна
- `date_crawled`, `date_created`, `last_seen` -  даты, которые также не понесут в себе особой пользы в обучении
- `postal_code` - слабо связанный со всеми фичами (+ таргетом) в датасете признак. Не несет практической значимости для предсказания целевого признака.

---

```python
data_autos_ml = data_autos.drop(['number_of_pictures',
                                 'date_crawled',
                                 'date_created',
                                 'last_seen',
                                 'postal_code',
                                 ], axis=1)

data_autos_ml.head()
```

```python
# Проверка датасета на пропущенные значения и дубликаты
print(f'Пропущенные значения: {data_autos_ml.isna().sum()}')
print(f'Количество дубликатов: {data_autos_ml.duplicated().sum()}')
print(f'Доля дубликатов от всех данных: {data_autos_ml.duplicated().sum() / len(data_autos_ml):%}')
```

```python
# Избавляемся от дубликатов
data_autos_ml.drop_duplicates(inplace=True)

# Проверка
print(f'Количество дубликатов после удаления: {data_autos_ml.duplicated().sum()}')
```

```python
RANDOM_STATE = 42
TEST_SIZE = 0.25

# Разбиваем фичи и таргет
X = data_autos_ml.drop('price', axis=1)
y = data_autos_ml['price']

# Разбиваем выборку на тренировочную и тестовую
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE)

print(f'X_train: {X_train.shape}')
print(f'X_test: {X_test.shape}')
print(f'y_train: {y_train.shape}')
print(f'y_test: {y_test.shape}')
display(X_train.head(1))
display(X_test.head(1))
display(y_train.head(1))
display(y_test.head(1))
```

### Пайплайн предобработки

```python
X_train.head(3)
```

```python
# Cоздаём списки с названиями признаков
ohe_columns = ['vehicle_type', 'gearbox', 'model', 'fuel_type', 'brand', 'repaired']
ord_columns = ['vehicle_type', 'gearbox', 'model', 'fuel_type', 'brand', 'repaired']
num_columns = ['registration_year', 'power', 'kilometer', 'registration_month']

# Cоздаём пайплайны для подготовки признаков 
ohe_pipe = Pipeline([
    ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])

ord_pipe = Pipeline([
    ('ord', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)),
    ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
])

# Cоздаём общий пайплайн для подготовки данных линейных моделей
data_preprocessor_lin = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('num', StandardScaler(), num_columns)
    ],
    remainder='passthrough'
)

# Cоздаём общий пайплайн для подготовки данных деревьев
data_preprocessor_tree = ColumnTransformer(
    [
        ('ord', ord_pipe, ord_columns),
        ('num', StandardScaler(), num_columns)
    ],
    remainder='passthrough'
)

# Cоздаём итоговый пайплайн: подготовка данных и базовая модель для линейных моделей
pipe_final_lin = Pipeline([
    ('preprocessor', data_preprocessor_lin),
    ('models', LinearRegression())
])

# Cоздаём итоговый пайплайн: подготовка данных и базовая модель для деревьев
pipe_final_tree = Pipeline([
    ('preprocessor', data_preprocessor_tree),
    ('models', RandomForestRegressor(random_state=RANDOM_STATE))
])

X_train_processed_lin = data_preprocessor_lin.fit_transform(X_train)
X_test_processed_lin = data_preprocessor_lin.transform(X_test)

X_train_processed_tree = data_preprocessor_tree.fit_transform(X_train)
X_test_processed_tree = data_preprocessor_tree.transform(X_test)
```

## Подбор гиперпараметров и обучение моделей

#### `LinearRegression`

```python
%%time

lin_reg = pipe_final_lin
lin_reg.fit(X_train, y_train)
```

```python
scores = cross_val_score(
    lin_reg,
    X_train,
    y_train,
    cv=5,
    n_jobs=-1,
    scoring='neg_root_mean_squared_error'
)

cv_score_lin_reg = sum(scores) / len(scores)
print('RMSE на кросс-валидации :', (-1) * cv_score_lin_reg)
```

```python
%%time

y_pred_lin_reg = lin_reg.predict(X_train)
print('RMSE на тренировочной выборке:', mean_squared_error(y_train, y_pred_lin_reg, squared=False))
```

---
- Качество предсказания `LinearRegression`:
    - RMSE на кросс-валидации: **2814.9993922170847**
    - RMSE на тренировочной выборке: **2809.2600167636288**

#### `RandomForestRegressor`

```python
# Подберем лучшие гиперпараметры

rf_params = {
    'models': [RandomForestRegressor(random_state=RANDOM_STATE)],
    'models__n_estimators': range(50, 100, 10),
    'models__max_depth': range(3, 8),
    'models__min_samples_leaf': range(5, 50, 5),
}

rf = RandomizedSearchCV(
    pipe_final_tree,
    rf_params,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1,
    n_iter=10,
    cv=5,
    verbose=5,
    random_state=RANDOM_STATE,
)

rf.fit(X_train, y_train)

# выведите лучшие гиперпараметры
print('Лучшая модель: \n', rf.best_estimator_)
# выведите лучшую метрику качества
print('RMSE на кросс-валидации', (-1) * rf.best_score_)
```
